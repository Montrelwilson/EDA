```python
#Question 1
result = NFL_log.groupby(['Against', 'Beneficiary'])['Yards'].sum().reset_index().sort_values(by='Yards', ascending=False)
print(result)
```

```python
#Question 2

penalties_2018 = NFL_log[NFL_log['Year'] == 2018]

# Group by 'Against' (the team that was penalized), count the penalties, and sort in descending order
team_penalty_counts = penalties_2018.groupby('Against').size().sort_values(ascending=False)
num_teams_penalized = team_penalty_counts.count()

team_penalty_counts, num_teams_penalized


```

```python
#Question 3
penalties_2015 = NFL_log[NFL_log['Year'] == 2015]

# Group by 'Against', count the penalties, and find the team with the most penalties
most_penalized_team_2015 = penalties_2015.groupby('Against').size().sort_values(ascending=False).head(1)

most_penalized_team_2015
```

```python
#Question 4

#Had previous issues with the below date so removing 
NFL_log_filtered = NFL_log[NFL_log['Date'] != '11/30/-0001']

NFL_log_filtered['Date'] = pd.to_datetime(NFL_log_filtered['Date'], errors='coerce')

# Defined the start and end dates for the period
start_date = pd.Timestamp('2009-09-01')
end_date = pd.Timestamp('2023-01-31')

# Filtered  the data between these dates
filtered_data = NFL_log_filtered[(NFL_log_filtered['Date'] >= start_date) & (NFL_log_filtered['Date'] <= end_date)]

# Filtered out entries where the player's name is 'Unnamed' or missing
valid_player_data = filtered_data[filtered_data['Player'].notna() & (filtered_data['Player'] != 'Unnamed')]

# Find the most penalized player with a valid name
most_penalized_player_valid = valid_player_data.groupby('Player').size().sort_values(ascending=False).head(1)

most_penalized_player_valid
```

```python
#Question 5
penalties_2022_df = NFL_log[NFL_log['Year'] == 2022]

parsons_penalties = penalties_2022_df[penalties_2022_df['Player'] == 'Micah Parsons']

# Count only the accepted penalties
accepted_penalties = parsons_penalties[parsons_penalties['Declined'] == 'No'].shape[0]

accepted_penalties
```

```python
#Question 6
penalties_2021_2022_df = NFL_log[(NFL_log['Year'] == 2021) | (NFL_log['Year'] == 2022)]

# Group by the penalty type and count the occurrences
penalty_counts = penalties_2021_2022_df.groupby('Penalty').size()

# Identifing the most common penalty
most_common_penalty = penalty_counts.sort_values(ascending=False).head(1)

# Count how many times "Roughing the Passer" was called
roughing_the_passer_count = penalty_counts.get("Roughing the Passer", 0)

most_common_penalty, roughing_the_passer_count
```

```python
#Question 1
NFL_Spreads=pd.read_csv("/work/spreads_scores-20231205-214226.csv")
display(NFL_Spreads)
```

```python
#Question 2
NFL_Spreads['Point_Differential'] = NFL_Spreads['score_home'] - NFL_Spreads['score_away']
display(NFL_Spreads)
```

```python
#Question 3
NFL_Spreads['Hwin'] = NFL_Spreads['score_home'] > NFL_Spreads['score_away']
NFL_Spreads['Awin'] = NFL_Spreads['score_away'] > NFL_Spreads['score_home']
print(NFL_Spreads.head())
```

```python
# Question 4
chiefs_games_2022 = NFL_Spreads[(NFL_Spreads['schedule_season'] == 2022) & 
((NFL_Spreads['team_home'] == 'Kansas City Chiefs') | 
(NFL_Spreads['team_away'] == 'Kansas City Chiefs'))]

chiefs_home_wins = chiefs_games_2022[(chiefs_games_2022['team_home'] == 'Kansas City Chiefs') & 
(chiefs_games_2022['Hwin'])].shape[0]

chiefs_away_wins = chiefs_games_2022[(chiefs_games_2022['team_away'] == 'Kansas City Chiefs') & 
(chiefs_games_2022['Awin'])].shape[0]

chiefs_home_wins, chiefs_away_wins

#17 games total
```
